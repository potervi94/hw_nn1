
Ваша задача методом градієнтного спуску побудувати модель для прогнозу ціни на діаманти

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module3/data/diamonds.csv",
                 index_col=0)

df.head()
```

Формула моделі
$$
          pred = coef*carat + intercept
$$
* coef - ціна за один карат
* intercept - витрати
  
  
<br>
<br>
Формула для ображунку похибки

$$
          loss = mean((pred - price)^2)
$$

* pred - ціна, спрогнозована моделю
* price - справжня ціна

```python
# імпорт модуля
import torch
from torch.optim import SGD

carat = df['carat'].to_numpy()
price = df['price'].to_numpy()

# вага в каратах
carat = torch.tensor(carat, dtype=torch.float32)

# справжня ціна
price = torch.tensor(price, dtype=torch.float32)
```

```python
# створіть тензори з градієнтом, оберіть довільне початкове значення, тип даних float32
coef = torch.tensor(1.0, dtype=torch.float32, requires_grad=True)
intercept = torch.tensor(1.0, dtype=torch.float32, requires_grad=True)

# вкажіть навчальний крок
learning_rate = 0.05  # замість 0.01

# визначення оптимізатора
optimizer = SGD([coef, intercept], lr=learning_rate)

# основний цикл
for t in range(5000):  # замість 2000
    # обрахуйте прогноз моделі за формулою наведеною вище
    pred = coef * carat + intercept

    # обрахуйте похибку між справжньою ціною та прогнозованою
    loss = torch.mean((pred - price) ** 2)

    # виведіть значення похибки для кожного сотого кроку
    if t % 100 == 0:
        print(f"Крок {t}: loss = {loss.item():.2f}, coef = {coef.item():.2f}, intercept = {intercept.item():.2f}")

    # обрахуйте градієнти по метриці
    loss.backward()

    # градієнтний спуск
    optimizer.step()
    optimizer.zero_grad()

    

```
